```python
%%R
library(dplyr)
library(googlesheets4)
library(ggthemes)

url = 'https://docs.google.com/spreadsheets/d/1_IWB5MBY7f44NgogU6-ysxNI--FknJw82DFG-3U3BFs/edit#gid=0'

wine <- read_sheet(url)

# Define a new color palette
custom_colors <- c("red", "blue", "green", "orange", "purple", "pink", "brown", "cyan", "magenta", "yellow")

wine %>%
  select(wine = Wine, variety = Variety, price = Price, borough = Borough) %>%
  ggplot(aes(x = variety, y = price, color = borough)) +
  geom_point(position = position_jitter(width = 0.2, height = 0.2), size = 3) +
  stat_summary(fun = median, geom = "point", shape = 17, size = 3, position = position_dodge(width = 0.2)) +
  labs(x = 'Variety', y = 'Price (in $)', title = '"120 Wine" prices across New York') +
  theme_clean() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1)) +
  scale_color_manual(values = custom_colors)
```

```python
%%R
library(dplyr)
library(googlesheets4)
library(ggthemes)

url = 'https://docs.google.com/spreadsheets/d/1_IWB5MBY7f44NgogU6-ysxNI--FknJw82DFG-3U3BFs/edit#gid=0'

wine <- read_sheet(url)

# Define a new color palette
custom_colors <- c("red", "blue", "green", "orange", "purple", "pink", "brown", "cyan", "magenta", "yellow")

wine %>%
  select(wine = Wine, variety = Variety, price = Price, Borough = Borough) %>%
  ggplot(aes(x = variety, y = price, color = Borough)) +
  geom_point(position = position_jitter(width = 0.2, height = 0.2), size = 3) +
  stat_summary(
    fun = median, 
    geom = "errorbar", 
    position = position_dodge(width = 0.2), 
    width = 0.2,
    color = "black",
    size = 1
  ) +
  labs(x = 'Variety', y = 'Price (in $)', title = '"120 Wine" prices across New York') +
  theme_clean() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1)) +
  scale_color_manual(values = custom_colors)
```
